## Mean - среднее значение

### среднее арифметическое значение по выборке
$$\bar{x} = \frac{1}{n}\sum_{i=1}^{n} x_i$$
### математическое ожидание
$$\mathbb{E}[X] = \sum_{k = 1}^{k} p_k \cdot v_k$$

```python
data = np.array([1, 2, 3, 4, 5, 6])
mean_np = np.mean(data)
mean_manual = data.sum()/data.size
print("Данные:      ", data)
print("np.mean:     ", mean_np)
print("Сумма/кол-во:", mean_manual)
```

```python
probs = np.array([0.1, 0.1, 0.5, 0.1, 0.1, 0.1])
math_expected = np.sum(data*probs)
math_expected_np = np.average(data, weights=probs)
print(f"E(x) = {math_expected.round(2)}")
print(f"E(x) from np = {math_expected_np.round(2)}")
```

## Mode = мода

### Самое частое значение

```python
samples = generator.integers(1, 7, size=200)
result = np.unique(samples, return_counts=True)
values = result[0]
counts = result[1]
mode_value = values[np.argmax(counts)]
for values, counts in zip(values, counts):
    print(f"грань {values}: выпала {counts:3d} раз  {'*' * (counts // 3)}")
print("\nМода (самая частая грань):", mode_value)
```
